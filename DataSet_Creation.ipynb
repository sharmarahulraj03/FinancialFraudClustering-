```python
sheet2_len = len(df2)
sheet2_yr = []
sheet2_tick = []
for i in range(0,sheet2_len):
    sheet2_year = df2.loc[i,'YEARA']
    sheet2_ticker = df2.loc[i,'Ticker']
    sheet2_yr.append(sheet2_year)
    sheet2_tick.append(sheet2_ticker)
```

```python
sheet1_len = len(df1)
data = []
ll=[]
for i in range(0,len(sheet2_yr)):
    x= sheet2_tick[i]
    y= sheet2_yr[i]
    counter_tick = 0
    for z in range(0,sheet1_len):
        if(counter_tick<=12):
            if(x==df1.loc[z,'Ticker']):
                if(y in df1.columns[2:]):
                    ll.append(df1.loc[z,y])
                    counter_tick+=1
        if(counter_tick>=13):
            data.append(ll)
            break
    if(counter_tick == 0):
      print(x)
      print(counter_tick)
```
